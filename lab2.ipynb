```python
import numpy as np
import matplotlib.pyplot as plt

import math
```

```python
### EX 1

PI = math.pi

def x(t, A = 1, f = 1, phi = PI / 3):
    # A = 1, f = 1, phi = pi/3
    return A * math.sin(2 * PI * f * t + phi)

def y(t, phi = PI / 3):
    # cos(t) = sin(t + pi/2) =(t=t-t/2) => cos(t-pi/2) = sin(t)
    return 1 * math.cos(2 * PI * 1 * t + phi - PI / 2)
```

```python
# extras din lab1
vals = np.linspace(start=0, stop=1, num=int(0.03/0.0005))

arr = []
for i in range(len(vals)):
    arr.append(x(vals[i]))

def print_plots(arr_f, frequency, f, phi):
    nr_plots = len(arr_f)
    for x in range(1, nr_plots + 1):
        print(x)
        ax = plt.subplot(nr_plots, 1, x)
        ax.set_title("Axa {}".format(x))
        ax.set_xlabel("Time")
        ax.set_ylabel("Amplitude")
        
        arr = []
        for i in range(len(vals)):
            if frequency == 0:
                arr.append(arr_f[x - 1](vals[i]))
            else:
                arr.append(arr_f[x - 1](vals[i], frequency, f, phi))
        ax.plot(vals, arr)
    
    plt.show()
    
print_plots(arr_f=[x, y], frequency=0, f = 1, phi = PI / 3)
```

```python
# Ex 2

f_list = [PI / 6, PI / 4, PI / 3, PI /2]

for f_index in range(len(f_list)):
    print_plots(arr_f=[x], frequency=0, f=1, phi = f_list[f_index])
```

```python
# partea cu SNR
# SNR = ||x||^2 / gama^2 * ||z||^2 , unde ||z|| este random.normal() - distributia gaussiana
vec = []
# 1000 valori
count = 1000
for i in range(count):
    vec.append(x(i))
    
def z(SNR):
    norm_x = np.linalg.norm(vec)

    # distributia gaussiana    
    gaussian_distribution = np.random.normal(size=(count, 1))
    
    # norma distributiei
    norm_z = np.linalg.norm(gaussian_distribution)
        
    return np.sqrt(norm_x / (SNR * norm_z))

SNR = [0.1, 1, 10, 100]

print("SNR values:")
for i in range(len(SNR)):
    print(z(SNR[i]))
```
