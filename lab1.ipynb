```python
import math

import matplotlib.pyplot as plt
import numpy
```

```python

# Ex 1

PI = math.pi

def x(t, frequency = 520):
    return math.cos(frequency * PI * t + PI / 3)

def y(t, frequency = 280):
    return math.cos(frequency * PI * t - PI / 3)

def z(t, frequency = 120):
    return math.cos(frequency * PI * t + PI / 3)
```

```python
# a, b
vals = numpy.linspace(start=0, stop=0.03, num=int(0.03/0.0005))

arr = []
for i in range(len(vals)):
    arr.append(x(vals[i]))

def print_plots(arr_f, frequency):
    nr_plots = len(arr_f)
    for x in range(1, nr_plots + 1):
        print(x)
        ax = plt.subplot(nr_plots, 1, x)
        ax.set_title("Axa {}".format(x))
        ax.set_xlabel("Time")
        ax.set_ylabel("Amplitude")
        
        arr = []
        for i in range(len(vals)):
            if frequency == 0:
                arr.append(arr_f[x - 1](vals[i]))
            else:
                arr.append(arr_f[x - 1](vals[i], frequency))
        ax.plot(vals, arr)
    
    plt.show()
    
print_plots(arr_f=[x, y, z], frequency=0)
```

```python
# c
print_plots(arr_f=[x, y, z], frequency=200)
```
