```python
import math

import matplotlib.pyplot as plt
import numpy as np
```

```python

# Ex 1

PI = math.pi

def x(t, frequency = 520):
    return math.cos(frequency * PI * t + PI / 3)

def y(t, frequency = 280):
    return math.cos(frequency * PI * t - PI / 3)

def z(t, frequency = 120):
    return math.cos(frequency * PI * t + PI / 3)
```

```python
# a, b
vals = np.linspace(start=0, stop=0.03, num=int(0.03/0.0005))

arr = []
for i in range(len(vals)):
    arr.append(x(vals[i]))

def print_plots(arr_f, frequency):
    nr_plots = len(arr_f)
    for x in range(1, nr_plots + 1):
        print(x)
        ax = plt.subplot(nr_plots, 1, x)
        ax.set_title("Axa {}".format(x))
        ax.set_xlabel("Time")
        ax.set_ylabel("Amplitude")
        
        arr = []
        for i in range(len(vals)):
            if frequency == 0:
                arr.append(arr_f[x - 1](vals[i]))
            else:
                arr.append(arr_f[x - 1](vals[i], frequency))
        ax.plot(vals, arr)
    
    plt.show()
    
print_plots(arr_f=[x, y, z], frequency=0)
```

```python
# c
print_plots(arr_f=[x, y, z], frequency=200)
```

```python
# 2(a) - Semnal sinusoidal de frecvență 400 Hz, 1600 esantioane
frecventa_a = 400
esantioane_a = 1600
t_a = np.linspace(0, 1, esantioane_a, endpoint=False)  # Intervalul de timp este 1 secunda
semnal_a = np.sin(2 * PI * frecventa_a * t_a)

plt.figure()
plt.subplot(321)
plt.plot(t_a, semnal_a)
plt.title("Semnal sinusoidal 400 Hz")

# 2(b) - Semnal sinusoidal de frecvență 800 Hz, 3 secunde
frecventa_b = 800
durata_b = 3  # Durata în secunde
esantioane_b = int(durata_b * 2000)  # 2000 esantioane pe secundă
t_b = np.linspace(0, durata_b, esantioane_b, endpoint=False)
semnal_b = np.sin(2 * PI * frecventa_b * t_b)

plt.subplot(322)
plt.plot(t_b, semnal_b)
plt.title("Semnal sinusoidal 800 Hz")

# 2(c) - Semnal de tip sawtooth de frecvență 240 Hz
frecventa_c = 240
esantioane_c = 2000
t_c = np.linspace(0, 1, esantioane_c, endpoint=False)
semnal_c = np.mod(t_c * frecventa_c, 1) * 2 - 1

plt.subplot(323)
plt.plot(t_c, semnal_c)
plt.title("Semnal sawtooth 240 Hz")

# 2(d) - Semnal de tip square de frecvență 300 Hz
frecventa_d = 300
t_d = np.linspace(0, 1, esantioane_c, endpoint=False)
semnal_d = np.sign(np.sin(2 * PI * frecventa_d * t_d))

plt.subplot(324)
plt.plot(t_d, semnal_d)
plt.title("Semnal square 300 Hz")

# 2(e) - Semnal 2D aleator
semnal_e = np.random.rand(128, 128)

plt.subplot(325)
plt.imshow(semnal_e, cmap='gray')
plt.title("Semnal 2D aleator")

# 2(f) - Semnal 2D la alegerea voastră (de exemplu, semnalul este o matrice de 128x128 de valori 1)
semnal_f = np.ones((128, 128))

plt.subplot(326)
plt.imshow(semnal_f, cmap='gray')
plt.title("Semnal 2D ales")

plt.tight_layout()
plt.show()
```
